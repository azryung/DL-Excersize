```python
#NAND게이트 / AND와 반대 설정이어야 하니까 마이너스로 한다고 생각하면 됨
#둘다 해당되면 0 False값을 반환한다~


import numpy as np

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7  
    tmp = np.sum(w*x) + b
    if tmp <=0:   #한계값이하면 0FALSE
        return 0  
    else :
        return 1  #한계값초과  1
    
if __name__ == '__main__':
    for xs in [(0,0),(1,0),(0,1),(1,1)]:
        y = NAND(xs[0], xs[1])
        print(str(xs)+" ->" + str(y))
```

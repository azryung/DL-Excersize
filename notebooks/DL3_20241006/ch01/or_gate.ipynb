```python
import numpy as np

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <=0:   #한계값을 0.2로 설정하고 싶으면 함수에서 마이너스 처리해서, 0이하값이 되는지 아닌지로 표현하는 거임~
        return 0  #한계값이하 0.2 이하이면 0
    else :
        return 1  #한계값초과 0.2 초과이면 1
    
if __name__ == '__main__':
    for xs in [(0,0),(1,0),(0,1),(1,1)]:
        y = OR(xs[0], xs[1])
        print(str(xs)+" ->" + str(y))
```

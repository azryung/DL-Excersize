```python
# coding: utf-8

import numpy as np
import matplotlib.pyplot as plt

import os, sys
sys.path.append(os.pardir)
print(os.getcwd())
current_dir = os.path.dirname(os.getcwd())
parent_dir = os.path.dirname(current_dir)
sys.path.append(parent_dir)
print(current_dir)

from dataset.mnist import load_mnist
from PIL import Image

(x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=False)
# #load_mnist 함수는 MNIST 손글씨 숫자 데이터를 불러오는 함수
# 데이터를 크게 두 덩어리로 나누어 반환(return)합니다.

# (x_train, t_train) : 훈련 데이터 (Training Data)

# x_train: 모델을 **학습(훈련)**시키기 위한 이미지 데이터입니다.

# t_train: x_train의 각 이미지에 해당하는 **정답(label)**입니다. (예: t_train[0]은 x_train[0] 이미지가 실제로 어떤 숫자인지 알려줍니다.)

# (x_test, t_test) : 테스트 데이터 (Test Data)

# x_test: 모델이 학습을 마친 후, 성능을 **시험(테스트)**해보기 위한 이미지 데이터입니다.

# t_test: x_test의 각 이미지에 해당하는 정답입니다.

# 함수의 인자(옵션) 의미
# flatten=True:

# 원래 MNIST 이미지는 (28, 28) 크기의 2차원 배열입니다.

# True로 설정하면, 이미지를 1줄로 쭉 펼쳐서 784개(28*28)의 숫자로 구성된 1차원 배열로 만들어 줍니다.

# 이것이 바로 img.shape가 (784,)로 나오고, img.reshape(28, 28)이 필요한 이유입니다.

# normalize=False:

# True로 설정하면 픽셀 값을 0.0 ~ 1.0 사이로 정규화합니다.

# False로 설정했기 때문에, 픽셀 값이 원래 값인 0 ~ 255 사이의 정수로 유지됩니다.



def img_show(img):
    plt.imshow(img, cmap = 'gray')
    plt.axis('off')  #축 숨기기 예쁘게 보이려구~
    plt.show()


img = x_train[1]
label = t_train[2]
print(label)

print(img.shape)  
img_re = img.reshape(28,28)  #원 해상도 맞추기 1차원 > 2차원 이미지로 변경
print(img_re.shape)  

img_show(img_re)

```

MNIST 데이터를 화면에 그려보자~

```python
#for문으로 반복해서 9 * 9 판으로 보이게~

# 1. 전체 그림판(figure)의 크기를 정합니다.

plt.figure(figsize = (12,12)) #가로세로 12인치
plt.subplots_adjust(hspace=0.3, wspace=0.3) #이미지 간의 가로세로 간격

# 2. for 반복문으로 9개의 이미지를 그립니다.
for i in range(81):
    # 3. 3x3 격자에서 i+1 번째 위치에 그림을 그릴 준비를 합니다.
    plt.subplot(9, 9, i + 1) # (9행, 9열, i+1번째)
    
    # 4. 데이터 가져오기 (이미지와 라벨의 인덱스 'i'를 일치시킵니다)
    img = x_train[i]
    label = t_train[i]
    
    # 5. 1차원 배열을 28x28 2차원 이미지로 변형
    img = img.reshape(28, 28) 
    
    # 6. 이미지 표시 (imshow) 및 축/제목 설정
    plt.imshow(img, cmap='gray')
    plt.title(f"Label : {label}") # 각 이미지 위에 정답(label) 표시
    plt.axis('off')             # 축 숨기기

# 7. 반복문이 끝난 후, 준비된 그림판을 한 번에 표시합니다.
plt.show()
```

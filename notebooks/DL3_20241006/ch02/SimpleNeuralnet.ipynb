## 입력층에서 1층으로 신호 전달

```python
# 입력값을 1층으로 보낼 수 있는 계산을 행렬이 해줌
# 넘파이에 행렬연산이 있어서 그걸 활용하는 거~~
# 교재 초기에 노드 레이어 배운게 있는데 그건가?

#Bias는 왜 들어가지?


X = np.array([1.0, 0.5])   #입력값
W1 = np.array([[0.1, 0.3, 0.5], [0.2,0.4,0.6]])  #가중치
B1 = np.array([0.1,0.2,0.3])  #bias

A1 = np.dot(X, W1) + B1

print('X=', X)
print('B1 = ', B1)
print('W1 = ', W1)

print('X * W1 + B1=', A1)

```

## 활성화 함수의 적용

```python
#활성화 함수? Activation?

Z1 = sigmoid(A1)
print('A1 = ', A1)
print('Z1 = ', Z1)
```

## 1층에서 2층으로 신호 전달

```python
#앞에서 1층 값을 활성화 함수를 통해서 
#Z1 1층에서 2층으로 갈때 입력값

#그림상으로 3개입력값이 2개 입력값으로 가야함 그래서 3행 - 2열

W2 = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])   #가중치
B2 = np.array([0.1,0.2])                #bias

A2 = np.dot(Z1, W2) + B2
print('Z1 * W2 + B1 = ', A2)
```

## 활성화 함수의 적용


```python
print('A2 = ', A2)

Z2 = sigmoid(A2)
print('Z2 = ', Z2)
```

## 2층에서 3층으로 신호 전달

```python
#최종적으로 2층 Z2 값을 3층으로 전달하기 위해서 최종 가중치를 적용하고
#기울기 소실을 막기 위해 bias B3 값을 적용함
#이때 2층은 2행 2열이고, 최종 신호 값은 1행 2열로 나옴

#Z2 입력 값

W3 = np.array([[0.2, 0.5], [0.3, 0.6]])
B3 = np.array([0.1,0.2])

A3 = np.dot(Z2, W3) + B3
print (f'W3 = {W3}')
print (f'B3 = {B3}')
print (f'Z2 * W3 + B3 = {A3}')



```

## 최종 출력층에 활성화 함수로 Softmax 적용

```python
print(f'A3= {A3}')

Y = softmax(A3)
print(f'Y={Y}')  #확률분포? 라서 두 값이 합해서 1 --> 결과가 0.626 -> 정답일 가능성이 높다~
```

## 입력층에서 1층으로 신호 전달

```python
# 입력값을 1층으로 보낼 수 있는 계산을 행렬이 해줌
# 넘파이에 행렬연산이 있어서 그걸 활용하는 거~~
# 교재 초기에 노드 레이어 배운게 있는데 그건가?

#Bias는 왜 들어가지?


X = np.array([1.0, 0.5])   #입력값
W1 = np.array([[0.1, 0.3, 0.5], [0.2,0.4,0.6]])  #가중치
B1 = np.array([0.1,0.2,0.3])  #bias

A1 = np.dot(X, W1) + B1

print('X=', X)
print('B1 = ', B1)
print('W1 = ', W1)

print('X * W1 + B1=', A1)

```

## 활성화 함수의 적용

```python
#활성화 함수? Activation?

def sigmoid(x):
    return 1 / (1+np.exp(-x))

Z1 = sigmoid(A1)
print('A1 = ', A1)
print('Z1 = ', Z1)
```

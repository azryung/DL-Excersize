```python
# coding: utf-8

import os, sys
from pathlib import Path
sys.path.append(os.pardir)
print(os.getcwd())
current_dir = os.path.dirname(os.getcwd())
print(current_dir)
os.chdir(current_dir)


import numpy as np
import pickle
from dataset.mnist import load_mnist
from common.functions import sigmoid, softmax


def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
    #이건 무슨 의미지
    return x_test, t_test

def init_network():
    print(os.getcwd())
    file_path = "C:/Users/EL90/OneDrive - (주)엘릭서/MS AI 8기/수업자료/workspace/DL-Excersize/notebooks/DL3_20241006/ch02/sample_weight.pkl"
    with open(file_path,"rb") as f:
        network = pickle.load(f)
    return network

def predict(network, x) :
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y


x, t = get_data()  #훈련함수? 만던그
network = init_network()

#배치크기를 정해줌
batch_size = 100
accuracy_count = 0

for i in range(0, len(x), batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis = 1)  #앞에서 나온 y 확률이 가장 높은 원소의 인텍스 숫자
    accuracy_count +=np.sum(p == t[i:i+batch_size])
    


print(f"정확도 : {str(float(accuracy_count)/len(x))}")

```
